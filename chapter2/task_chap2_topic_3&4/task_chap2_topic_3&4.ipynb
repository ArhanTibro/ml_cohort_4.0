```python
import pandas as pd
import warnings
warnings.filterwarnings('ignore')
housing=pd.read_csv("Housing.csv")
housing.info()
```

```python
housing.describe()
```

```python
#an example function to show the use case of numpy on dataset
def split_train_test(data,test_ratio):
    np.random.seed(42)
    shuffled=np.random.permutation(len(data))
    print(shuffled)
    test_set_size=int(len(data)*test_ratio)
    test_indices=shuffled[:test_set_size]
    train_indices=shuffled[test_set_size:]
    return data.iloc[train_indices],data.iloc[test_indices]
```

```python
#line plot
import matplotlib.pyplot as plt
housing.area.plot(kind='line',color='g',label='area',linewidth=1,alpha=0.5,grid=True,linestyle=':')
housing.price.plot(color='r',label='price',linewidth=1,alpha=0.5,grid=True,linestyle='-.')
plt.legend(loc='upper right') #puts label into plot
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.title('line plot')
plt.show()
```

```python
#scatter plot

housing.plot(kind='scatter',x='area',y='price',alpha=0.5,color='red')
plt.xlabel('Area')
plt.ylabel('Price')
plt.title('area price scatter plot')
```

```python
#Histogram plot
#bins=number of bars in the figure

housing.price.plot(kind='hist',bins=50,figsize=(12,12))
plt.show()
```
