```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df=pd.read_csv("Books_Data_Clean.csv")
df.head()
```

```python
df.describe()
```

```python
#fix the negative value in publishing year
df=df[df['Publishing Year']>1900]
df.describe()
```

```python
#check if any null value exists 
df.isna().sum()
```

```python
#drop rows where books doesn't have names
df.dropna(subset="Book Name",inplace=True) 
df.isna().sum() 
```

```python
#check if there any duplicate value
df.duplicated().sum()
```

```python
df.keys()
```

```python
#check the unique values
df.nunique()
```

```python
#use histogram  for the publishing year
plt.hist(df['Publishing Year'],bins=20)
plt.xlabel('publishing year')
plt.ylabel('Frequency')
plt.title("Distribution of publishing year")
plt.show()
```

```python
#see number of books for each genre
df['genre'].value_counts().plot(kind='bar')
plt.xlabel("Genre")
plt.ylabel("Number of books")
```

```python
#see authors bookwise average rating
df.groupby('Author')['Book_average_rating'].mean().sort_values(ascending=False)
```

```python
#visualize Book_ratings_count according to genre 
sns.boxenplot(x='genre',y='Book_ratings_count',data=df)
plt.xlabel("Genre")
plt.ylabel("books ratings count")
plt.show()
```

```python
#see correlation between sale price and unit sold using scatter plot
plt.scatter(x=df['sale price'],y=df['units sold'])
#you can add xlabel , ylabel and title
plt.show()
```

```python
#see the value count of the languages
lc=df['language_code'].value_counts().sort_values(ascending=True)
lc
```

```python
#see the language distribution in pie chart
plt.pie(lc,labels=lc.index,startangle=90,autopct="%1.1f%%")
plt.show()
```

```python
#see the total revenue of the publishers
df.groupby('Publisher ')['publisher revenue'].sum().sort_values(ascending=False)
```

```python
#group the data by author rating and see the average book rating count for each author rating
df.groupby("Author_Rating")["Book_ratings_count"].mean()
```

```python
#Book Average Rating vs Book Ratings Count in scatter plot
plt.scatter(df['Book_average_rating'],df['Book_ratings_count'])
plt.xlabel("Book Average Rating")
plt.ylabel("Book Ratings Count")
plt.show()
```

```python
#see total gross sales by each author
total_gross_sales=df.groupby("Author")["gross sales"].sum()
total_gross_sales.head()
```

```python
total_gross_sales.sort_values(ascending=False).head(10).plot(kind='bar')
plt.xlabel("Author")
plt.ylabel("Total gross sales")
plt.show()
```

```python
#see boxplot for unit sold for each author rating
sns.boxenplot(x="Author_Rating",y="units sold",data=df)
plt.xlabel("Author Rating")
plt.ylabel("Units sold")
plt.show()
```

```python
#see a line graph of total units sold for each publishing year
df.groupby("Publishing Year")["units sold"].sum().plot(kind="line",marker="o")
plt.xlabel("publishing year")
plt.ylabel("units sold")
plt.title("total units sold for each publishing year")
plt.show()
```
