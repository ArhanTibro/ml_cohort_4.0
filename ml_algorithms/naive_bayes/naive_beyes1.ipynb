```python
import pandas as pd
df=pd.read_csv("titanic.csv")
df.head()
```

```python
#remove unnecessary columns
df.drop(['PassengerId','Name','SibSp','Parch','Ticket','Cabin','Embarked'],axis='columns',inplace=True)
df.head()
```

```python
#separate target and features
target=df.Survived
inputs=df.drop('Survived',axis='columns')

#convert the sex column into boolean format in another columns
dummies=pd.get_dummies(inputs.Sex)
dummies.head()
```

```python
inputs=pd.concat([inputs,dummies],axis='columns')
inputs.drop('Sex',axis='columns',inplace=True)
inputs.head()
```

```python
#check if there are any nan values in any columns
inputs.columns[inputs.isna().any()]
```

```python
#fill the nan values with mean value of the column
inputs.Age=inputs.Age.fillna(inputs.Age.mean())
inputs.Age.head(6)
```

```python
#train test split
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(inputs,target,test_size=0.2)
```

```python
#gaussianNB can be used where normal distribution exists
from sklearn.naive_bayes import GaussianNB
model=GaussianNB()
model.fit(x_train,y_train)
y_predict=model.predict(x_test)
model.predict_proba(x_test)
```
